L'ancien choixpeau magique classait les élèves dans les différentes maisons en fonction de certaines de leurs caractéristiques psychologiques. Le choix se faisait de la manière suivante : 
- Gryffondor : si l'élève privilégiait le courage, la bravoure et la défense des valeurs,
- Serdaigle : si l'élève valorisait l'intelligence, la sagesse et la créativité,
- Poufsouffle : si l'élève mettait en avant la loyauté, le travail acharné et la bienveillance,
- Serpentard : si l'élève accordait de l’importance à l’ambition, la ruse et la détermination.

La professeur Sibylle Trelawney propose alors un questionnaire à 4 questions seulement :
- Question 1 : À quel point aimez-vous prendre des risques pour défendre une cause juste ?
- Question 2 : Comment évaluez-vous votre curiosité intellectuelle et votre soif d’apprendre ?
- Question 3 : Quelle importance accordez-vous à la loyauté envers vos amis et votre famille ?
- Question 4 : À quel point êtes-vous ambitieux et déterminé à atteindre vos objectifs personnels ? 

Les élèves doivent alors répondre à chaque question avec un score compris entre 1 (pas du tout) et 10 (complètement) représentant sa réponse. Le choix se fait en affectant l'élève à la maison :
- Gryffondor : si l'élève a mis le meilleur score à la Question 1,
- Serdaigle : si l'élève a mis le meilleur score à la Question 2,
- Poufsouffle : si l'élève a mis le meilleur score à la Question 3,
- Serpentard : si l'élève a mis le meilleur score à la Question 4.

Si un élève a mis le meilleur score à plusieurs questions, on l'affecte à n'importe quelle maison à laquelle il a mis le meilleur score.


**Remarque :** Même si la magie ne peut pas répartir les élèves, elle peut agrandir ou rétrécir sans souci les chambres et les salles de classe donc il n'y a pas de contrainte liée au nombre d'élèves affectés à chaque maison. Il vaut mieux répartir correctement chaque élève sans affecter le même nombre d'élèves à chaque maison plutôt que mal répartir les élèves de manière équitable !

Concrètement, on décide de stocker les réponses au questionnaire dans un tableau de chaînes de caractères et d'entiers contenant successivement un élève (sous forme d'une chaîne de caractères), et les quatres réponses aux questions sous forme d'entiers. Le nom de chaque élève se trouve à la case d'indice `[5*i]`, les réponses aux questions 1, 2, 3 et 4 se trouvent respectivement aux cases `[5 *i + 1]`, `[5 *i + 2]`, `[5 *i + 3]` et `[5 *i + 4]`. Un tel tableau est appelé *tableau de réponses*. On suppose qu'un tableau de réponses est toujours correctement construit (une chaîne de caractères pour chaque élève et 4 entiers entre 1 et 10 pour les réponses).

### Question 1

Définissez la fonction `nombre_eleves` prenant en paramètre un tableau de réponses et retournant le nombre d'élèves ayant répondu au questionnaire. Définissez également une fonction de tests unitaires.

````python
def nombre_eleves(tab):
    """
    tab(tableau en paramètre)
    renvoie le nombre d'élèves
    BUT: le programme compte le nombre d'élève ayant répondu au questionnaire sachant que dans le tableau la position entre un élève et un autre différencie de 5.
    """
    i=0                                   #Initialisation d'une variable
    nb_eleves=0                           #Initialisation d'une variable
    while i<len(tab):                     #Boucle qui permet de parcourir dans le tableau
        nb_eleves+=1                      #Incrémentation du nombre d'élève de 1
        i=i+5                             #Incrémentation de l'indice de 5 (récupérer seulement les noms)
    return nb_eleves                      #Renvoie le nombre d'élève

def test_nombre_eleves():                 #Foncion test qui teste la fonction nombre_eleves
    assert nombre_eleves(reponses)==3
    print("Test de la fonction nombre_eleves : OK")


test_nombre_eleves()                      #Appel de la fonction test
nombre_eleves(reponses)                   #Appel de la fonction nombre_eleves
````

Définissez la fonction `maison` prenant en paramètre un tableau de réponses ainsi que l'indice d'un élève. La fonction retourne la maison correspondant à l'affectation de cet élève. Ainsi, l'appel de la fonction maison avec le tableau de réponses donné en exemple doit retourner : 
- `"Gryffondor"` avec l'indice `0` (correspondant à Harry Potter),
- `"Poufsouffle"` avec l'indice `5` (correspondant à Cedric Diggory),
- `"Serpentard"` avec l'indice `10` (correspondant à Drago Malefoy).

En effet, Harry Potter a mis le meilleur score à la question 1 et doit donc être affecté à Gryffondor. Cedric Diggory a mis le meilleur score à la question 3 et doit être affecté à la maison Poufsouffle. Finalement, Drago a mis le meilleur score à la question 4 et doit donc être affecté à la maison Serpentard.

Définissez également une fonction de tests unitaires pour la fonction `maison`.

````python
from random import randint         #Importation du module random
def maison(tab,indice):            
    """
    tab(tableau)--> contient toutes les réponses des élèves
    indice(entier)--> correspond à la position d'un éléève dans le tableau
    renvoie la maison à laquelle est affecté l'élève
    """
    len=indice+5                   #Initialisation d'une varible pour avoir la limite de parcours
    pos_final=indice               #Initialisation, d'une variable pour avoir la l'indice de départ (sera utiliser dans la fonction vers la fin)
    indice+=1                      #Incrémentation de l'indice pour ne pas tenir compte du nom de l'élève
    max=0                          #Initialisation d'une variable et stockage de la première réponse de l'élève
    pos=0                          #Initialisation d'une varible pour retenir la position du maximum
    while indice<len:              #Boucle pour parcourir le tableau de réponses
        if tab[indice]>max:        #Condition pour vérifier si la valeur à l'indice dans le tableau est supérieur au maximum déjà connus
            max=tab[indice]        #Remplacement du maximum
            pos=indice             #Remplacement de la position du maximum
        elif tab[indice]==max:     #Condition si égalité entre max et la valeur à l'indice dans le tableau
            choix=randint(0,1)     #Choix aléatoire entre 0 et 1
            if choix==1:           #Condition si le choix vaut 1
                max=tab[indice]    #Remplacement du maximum
                pos=indice         #Remplacement de la postion du maximum
        indice+=1                  #Incrémentation de l'indice de 1
    if pos==pos_final+1:           #Condition si la postion correspond à la première réponse
        return "Gryffondor"        #Renvoie Gryffondor
    elif pos==pos_final+2:         #Condition si la postion correspond à la deuxième réponse
        return "Serdaigle"         #Renvoie Serdaigle
    elif pos==pos_final+3:         #Condition si la postion correspond à la troisième réponse
        return "Poufsouffle"       #Renvoie Poufsouffle
    else:                          #Condition sinon la postion correspond à la quatrième (dernière) réponse
        return "Serpentard"        #Renvoie Serpentard

def test_maison():                 #Foncion test qui teste la fonction maison                               
    assert maison(reponses,0)=="Gryffondor"
    assert maison(reponses,5)=="Poufsouffle"
    assert maison(reponses,10)=="Serpentard"
    print("Test de la fonction lecture_reponses : Ok")


test_maison()                      #Appel de la fonction test
print(maison(reponses,0))          #Affichage du renvoie de l'appel de fonction maison 
print(maison(reponses,5))          #Affichage du renvoie de l'appel de fonction maison 
print(maison(reponses,10))         #Affichage du renvoie de l'appel de fonction maison 
````

````python
def nb_erreurs(dico1,dico2):     
    """
    dico1(dictionnaire en paramètre)-->contient les noms de chaque élèves et leur maison
    dico2(dictionnaire en paramètre)-->contient les noms de chaque élèves et leur maison
    La fonction compare la maison de chaque élève dans les deux dictionnaires et s'il y a des différences alors renvoie le pourcentages d'érreurs (différence)
    """
    tab_dico1_cle=list(dico1)                            #Stockage des clés du dico1
    tab_dico1_valeur=list(dico1.values())                #Stockage des valeur du dico1
    tab_dico2_cle=list(dico2)                            #Stockage des clés du dico2
    tab_dico2_valeur=list(dico2.values())                #Stockage des valeur du dico2
    nb=0                                                 #Initialisaion d'une variable
    i=0                                                  #Initialisaion d'une variable
    while i<len(tab_dico1_cle):                          #Boucle pour parcourir tous les éléments du tabeau qui contient les clés du dico1
        j=0                                              #Initialisaion d'une variable
        while j<len(tab_dico2_cle):                      #Boucle pour parcourir tous les éléments du tabeau qui contient les clés du dico2
            if tab_dico1_cle[i]==tab_dico2_cle[j] and tab_dico1_valeur[i]==tab_dico2_valeur[j]:           #Condition si les clés et les valeurs dans les deux tableaux sont identiques
                nb+=1                                    #Incrémentation du nombre de similarité de 1
            j+=1                                         #Incrémentation de l'indice de 1
        i+=1                                             #Incrémentation de l'indice de 1
    resultat=len(tab_dico1_cle)-nb                       #Calcul du nombre d'erreurs
    return resultat                                      #Renvoie le nombre d'erreurs


def test_nb_erreurs():                                   #Foncion test qui teste la fonction nb_erreurs                               
    assert nb_erreurs({"Harry Potter": "Gryffondor", "Cedric Diggory": "Poufsouffle", "Drago Malefoy": "Serpentard"},{"Cedric Diggory": "Serdaigle", "Drago Malefoy": "Gryffondor", "Harry Potter": "Gryffondor"})==2
    print("Test de la fonction nb_erreurs : Ok")

fichier=open('affectation_premiere_annee.json')
fichier=load(fichier)

test_nb_erreurs()                                        #Appel de la fonction test
print(str((nb_erreurs({"Harry Potter": "Gryffondor", "Cedric Diggory": "Poufsouffle", "Drago Malefoy": "Serpentard"},{"Cedric Diggory": "Serdaigle", "Drago Malefoy": "Gryffondor", "Harry Potter": "Gryffondor"})/3)*100)+"%")        #Affichage du pourcentage d'erreur pour l'exemple demandé
print(nb_erreurs(repartition(lecture_reponses('questionnaire_premiere_annee.txt')),fichier))                                                                                                                                           #Affichage du nombre d'erreurs (valeur variante car choix au hasard)
print(str((nb_erreurs(repartition(lecture_reponses('questionnaire_premiere_annee.txt')),fichier)/len(fichier))*100)+"%")                                                                                                               #Affichage du poucentage d'erreurs (valeur variante car choix au hasard) 
````

````python
tab_eleves=eleves(lecture_reponses('questionnaire_premiere_annee.txt'))    #Tableau contenant les noms de tous les élèves
tab_resultats=[]                                                           #Nouveau tableau
j=0                                                                        #Initialisation d'une variable
while j<100:                                                               #Boucle pour ajouter 100 éléments dans le nouveau tableau
    dico_aleatoire={}                                                      #Nouveau dictionnaire
    i=0                                                                    #Initialisation d'une variable
    while i<len(tab_eleves):                                               #Boucle pour parcourir le tableau des noms des élèves
        choix=randint(1,4)                                                 #Choix aléatoire entre 1 et 4
        if choix==1:                                                       #Condition si le choix vaut 1
            dico_aleatoire[tab_eleves[i]]="Gryffondor"                     #Ajoute dans le dictionnaire en clé l'élève à la position i dans le tableau des noms élèves et comme valeur Gryffondor
        elif choix==2:                                                     #Condition si le choix vaut 2
            dico_aleatoire[tab_eleves[i]]="Serdaigle"                      #Ajoute dans le dictionnaire en clé l'élève à la position i dans le tableau des noms élèves et comme valeur Serdaigle
        elif choix==3:                                                     #Condition si le choix vaut 3
            dico_aleatoire[tab_eleves[i]]="Poufsouffle"                    #Ajoute dans le dictionnaire en clé l'élève à la position i dans le tableau des noms élèves et comme valeur Poufsouffle
        else :                                                             #Condition si le choix vaut autre (4)
            dico_aleatoire[tab_eleves[i]]="Serpentard"                     #Ajoute dans le dictionnaire en clé l'élève à la position i dans le tableau des noms élèves et comme valeur Serpentard
        i+=1                                                               #Incrémentation de l'indice de 1
    erreur=((nb_erreurs(repartition(lecture_reponses('questionnaire_premiere_annee.txt')),dico_aleatoire))/len(dico_aleatoire))*100          #Calcul pour avoir le nombre d'erreurs (valeur variante car choix au hasard) avec les élèements du fichier
    tab_resultats.append(erreur)                                           #Ajout dans le nouveau tableau le nombre d'erreur
    j+=1                                                                   #Incrémentation du nombre de répartition de 1

i=0                                                                        #Initialisation d'une variable
somme=0                                                                    #Initialisation d'une somme
while i<len(tab_resultats):                                                #Boucle pour parcourir le tableau contenant les nombres d'erreurs pour chaque tentative
    somme+=tab_resultats[i]                                                #Calcul de chaque valeur dans le tableau
    i+=1                                                                   #Incrémentation de l'indice de 1
moyenne=somme/len(tab_resultats)                                           #Calcul de la moyenne d'erreur total
print(moyenne)                                                             #Affichage de cette moyenne
````
